# Filtering on Array Fields

```python
from pymongo import MongoClient
import pprint
from IPython.display import clear_output
```

```python
##Movies Scratch Collection
client = MongoClient('mongodb+srv://analytics:analytics-password@cluster0.sp8wc.mongodb.net/movies_scratch?retryWrites=true&w=majority')
print(client.Cluster0)
```

## Exact Match Filtering
#### Finds documents whose languages array just includes "Korean" and "English" records with exact order

```python
filter = {
    'languages':['Korean', 'English']
}
projection = {
    '_id': 0,
    'title':1,
    'languages':1
}
```

```python
pprint.pprint(list(client.Cluster0.movies_scratch.find(filter, projection)))
```

## Coverage Filtering
#### Finds all documents that includes any of Korean and English in languages array.

```python
filter = {
    'languages':{'$all':['Korean', 'English']}
}
projection = {
    '_id': 0,
    'title':1,
    'languages':1
}
```

```python
pprint.pprint(list(client.Cluster0.movies_scratch.find(filter, projection)))
```

## Another Filtering :)
#### Finds all documents with the first element is "Korean"

```python
filter = {
    'languages.0': 'Korean'
}
projection = {
    '_id': 0,
    'title':1,
    'languages':1
}
```

```python
pprint.pprint(list(client.Cluster0.movies_scratch.find(filter, projection)))
```
